```r
library(Seurat)
library(Matrix)
options(future.globals.maxSize = Inf) 
set.seed(666)
rds_dir="/diskmnt/Projects/myeloma_scRNA_analysis/MMY_IRD/revision/individual_samples/rds/"
mtx_dir = "/diskmnt/Projects/myeloma_scRNA_analysis/MMY_IRD/revision/individual_samples/mtx/"
setwd(mtx_dir)
```

```r
di_table <- read.table("/diskmnt/Projects/myeloma_scRNA_analysis/MMY_IRD/revision/sample_clin_data/sample_table_DI.txt", sep="\t", header=T)

di_table$rds_path <- paste0(rds_dir,di_table$rds)

existing_files_table <- di_table[which(file.exists(di_table$rds_path)),]
```

```r
existing_files_table
```

```r
for(i in 1:nrow(existing_files_table)){
    s=existing_files_table[i, 'Sample']
    print(s)
    dir.create(s)
    
    objpath <- existing_files_table[i, 'rds_path']
    obj <- readRDS(objpath)

    counts <- GetAssayData(obj, assay = "RNA", layer = "counts")
    print(max(counts))
    print(dim(counts))
    
    writeMM(counts, file = paste0(s,"/counts.mtx"))
    write.table(rownames(counts), file = paste0(s,"/genes.tsv"), quote = FALSE, col.names = FALSE, row.names = FALSE)
    write.table(colnames(counts), file = paste0(s,"/barcodes.tsv") , quote = FALSE, col.names = FALSE, row.names = FALSE)

    meta <- obj@meta.data
    write.table(meta, file=paste0(s ,"/metadata.tsv"), quote = FALSE, col.names = TRUE, row.names = TRUE)
}
```
